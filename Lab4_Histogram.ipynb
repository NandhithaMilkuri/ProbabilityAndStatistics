```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
file_path = "Titanic_Dataset.csv"

df = pd.read_csv(file_path)
```

```python
df.rename(columns={
    "Passenger ID": "PassengerID",
    "Gender": "Sex",
    "Class": "Pclass",
    "Siblings/Spouses Aboard": "SibSp",
    "Parents/Children Aboard": "Parch"
}, inplace=True)
```

```python
df["Pclass"] = df["Pclass"].map({"1st": 1, "2nd": 2, "3rd": 3})

sns.set_style("whitegrid")
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['Age'], bins=15, kde=True, color='skyblue')
plt.xlabel("Age")
plt.ylabel("Count")
plt.title("Distribution of Passenger Ages")
plt.axvline(df['Age'].median(), color='red', linestyle='dashed', label='Median Age')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='Age', y='Fare', hue='Survived', alpha=0.7, palette={0: 'red', 1: 'green'})
plt.xlabel("Age")
plt.ylabel("Fare Paid")
plt.title("Fare vs. Age with Survival Status")
sns.regplot(data=df, x='Age', y='Fare', scatter=False, color='black')
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='Pclass', hue='Pclass', palette="pastel", legend=False)
plt.xlabel("Passenger Class")
plt.ylabel("Number of Passengers")
plt.title("Passenger Count by Class")
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='Pclass', hue='Survived', palette={0: 'red', 1: 'green'})
plt.xlabel("Passenger Class")
plt.ylabel("Count")
plt.title("Survival Count by Passenger Class")
plt.legend(title="Survived", labels=["No", "Yes"])
plt.show()
```
